```python
import numpy as np
from scipy.stats import norm

n = 250
x = 70
p_hat = x/n     # 0.28

z_norm = norm(loc=0, scale=1)

# 90% 신뢰구간
ll_90 = p_hat - z_norm.ppf(0.95)*np.sqrt(p_hat*(1-p_hat)/n)
ul_90 = p_hat + z_norm.ppf(0.95)*np.sqrt(p_hat*(1-p_hat)/n)

# 95% 신뢰구간
ll_95 = p_hat - z_norm.ppf(0.975)*np.sqrt(p_hat*(1-p_hat)/n)
ul_95 = p_hat + z_norm.ppf(0.975)*np.sqrt(p_hat*(1-p_hat)/n)

print("The point estimator of the proportion is {:.4f}".format(p_hat))
print("90% confidence interval of the proportion is ({:.4f}, {:.4f})".format(ll_90, ul_90))
print("95% confidence interval of the proportion is ({:.4f}, {:.4f})".format(ll_95, ul_95))
print()
print("비교, 차이점")
print(" - 두 신뢰구간 모두 중심은 0.28이다. -> 표본에서 얻은 비율 자체는 변하지 않는다.")
print(" - 90% 신뢰구간(0.2333, 0.3267) 보다 95% 신뢰구간(0.2243, 0.3357)이 더 넓다.")
print(" - 95% 신뢰수준은, 모집단의 실제 비율을 포함할 확률이 높다.")
```
